```python
###########################
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
 
###########################
# 1.과거의 데이터를 준비합니다.
PL = pd.read_csv("/home/smsung/Downloads/MilliCar-SUMO/Test/Test_data_220120/cv_s60_i0.0_k0.0_a0.0_l_S_Highway.txt",sep='\t', names=['seq','speed','IR','Pathloss','dist','WA','WC','SC'])

# 원핫인코딩
PL = pd.get_dummies(PL)

# 종속변수 x, 독립변수 y
x = PL[['seq','speed','IR','dist','WA','WC']]
y = PL[['Pathloss']]
print(x.shape, y.shape)
 
###########################
# 2. 모델의 구조를 만듭니다
X = tf.keras.layers.Input(shape=[6])
H = tf.keras.layers.Dense(24, activation="tanh")(X)
H = tf.keras.layers.Dense(12, activation="tanh")(H)
H = tf.keras.layers.Dense(8, activation="tanh")(H)
Y = tf.keras.layers.Dense(1, activation='sigmoid')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mean_squared_error',
              metrics=['accuracy'])
 
# 모델 구조 확인
model.summary()
 
###########################
# 3.데이터로 모델을 학습(FIT)합니다.
model.fit(x, y, epochs=100)
 
###########################
# 4. 모델을 이용합니다
print(model.predict(x[:7]))
print(y[:7])
```
