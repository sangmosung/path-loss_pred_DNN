```python
import tensorflow as tf
import pandas as pd
```

```python
df = pd.read_csv("/home/smsung/Downloads/MilliCar-SUMO/Test/Test_data_220119/cv_s60_i0.0_k0.0_a0.0_l_S_Highway.txt",sep='\t', names=['seq','speed','IR','Pathloss','dist','WA','WC','SC'])

data = df.drop(columns='SC')
```

```python
sm = data[['dist']]
sung = data[['Pathloss']]

print(sm.shape, sung.shape)
```

```python
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)

model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```python
model.fit(sm, sung, epochs=1000)
```

```python
model.predict(sm)
```
